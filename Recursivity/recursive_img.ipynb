```python
# Read image file
import matplotlib.pyplot as plt

# https://github.com/lgs5728/Introduccion_Programacion_Python/blob/master/fig02.jpg
# img = plt.imread("https://github.com/lgs5728/Introduccion_Programacion_Python/blob/master/fig02.jpg", format="jpg")
img = plt.imread("fig02.jpg", format = "jpg")
```

```python
# Plot image
plt.imshow(img)
plt.show()

plt.imshow(img, cmap="gray")
plt.show()
```

```python
# Generate histogram
plt.hist(img)
```

```python
# Generate array with numpy
import numpy as np

# Reshape image object
ss = img.shape
v = np.reshape(img, np.prod(ss))

# Plot histogram
plt.hist(v)
plt.show()

# Modify number of bins
plt.hist(v, bins=20)
plt.show()
```

```python
# Threshold image
v = 255 * (v > 225)
# Reshape
img_th = np.reshape(v, ss)
# Plot
plt.imshow(img_th, cmap="gray")
```

```python
# Extract regions of interest
i0 = 250
j0 = 330
# Be sure is not an empty pixel
print(img_th[i0, j0])

x=[]
y=[]

def roi(i,j):
    if img_th[i,j] == 255:
        x.append(i)
        y.append(j)
        img_th[i,j] = 0
        for n in [i-1, i, i+1]:
            for m in [j-1, j, j+1]:
                if i != n and j != m:
                    roi(n,m)

# Call function
roi(i0,j0)

# Create new image
img_roi = np.zeros(ss)
# Add ROI
nx = len(x)
for i in range(nx):
    img_roi[x[i],y[i]] = 255
# Plot ROI
plt.imshow(img_roi, cmap="gray")
```
